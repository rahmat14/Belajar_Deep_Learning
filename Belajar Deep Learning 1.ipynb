```python
import numpy as np
import matplotlib.pylab as plt
```

> Langkah 2 : import package arsitektur deep learning
> - keras : package yang menyediakan framework deep learning, link : https://keras.io/

```python
import keras
from keras.models import Model
from keras.models import Sequential
from keras.layers import Input
from keras.layers import Activation
from keras.layers import Dense
from keras.optimizers import SGD
```

> Langkah 3 : membuat contoh data training menggunakan rumus matematika
> - x_train : membuat data dari -30 sampai 30 dengan jarak 0.25, $\color{black}x=(-10,-9.75, \dots ,9.75)$
> - y_train : menggunakan rumus matematika $\color{black}y = \sqrt{x^2+1}$

```python
x_train = np.arange(-10, 10, 0.25)
y_train = np.sqrt((x_train**2)+1)
```

> Langkah 4 : membangun arsitektur artificial neural network yang terdiri,
> - 1 Input Node
> - 8 node pada Hidden Layer 1 dengan ReLU activation
> - 4 node pada Hidden Layer 2 dengan ReLU activation
> - 1 Output node dengan Linear activation
> - link kumpulan fungsi activation : https://en.wikipedia.org/wiki/Activation_function
> - Berikut ilustrasi dari arsitektur neural network
> <a href="https://ibb.co/e5eMAK"><img src="https://preview.ibb.co/jhR1AK/4.jpg" alt="4" border="0"></a>


```python
inputs = Input(shape=(1,))
hidden_layer_1 = Dense(128, activation='relu')(inputs)
hidden_layer_2 = Dense(64, activation='relu')(hidden_layer_1)
outputs = Dense(1, activation='linear')(hidden_layer_2)
model_1 = Model(inputs, outputs)
model_1.compile(loss = 'mean_squared_error', 
              optimizer = SGD(lr=0.01), 
              metrics = ['accuracy'])

model_2 = Model(inputs, outputs)
model_2.compile(loss='mean_squared_error', 
              optimizer='rmsprop', 
              metrics=['accuracy'])
```

> Langkah 5 : setelah membangun arsitektur neural network
>             maka komputer akan melatihnya,
>             penjelasan kode program ada di link : https://keras.io/

```python
model_1.fit(x_train, y_train, 
          batch_size = 30, 
          epochs = 100, 
          verbose = 1)
```

```python
model_2.fit(x_train, y_train, 
          batch_size = 10, 
          epochs = 100, 
          verbose = 1)
```

> Langkah 6 : menyimpan model prediksi hasil dari proses learning arsitektur neural network

```python
predict = model_1.predict(np.array([4]))
print('f(26) = ', predict)
```

```python
predict = model_2.predict(np.array([26]))
print('f(26) = ', predict)
```

> kode bahasa pemprogramman python Lengkap

```python
import numpy as np
import matplotlib.pylab as plt

import keras
from keras.models import Model
from keras.models import Sequential
from keras.layers import Input
from keras.layers import Activation
from keras.layers import Dense
from keras.optimizers import SGD

x_train = np.arange(-10, 10, 0.25)
y_train = np.sqrt((x_train**2)+1)

inputs = Input(shape=(1,))
hidden_layer_1 = Dense(128, activation='relu')(inputs)
hidden_layer_2 = Dense(64, activation='relu')(hidden_layer_1)
outputs = Dense(1, activation='linear')(hidden_layer_2)
model_1 = Model(inputs, outputs)
model_2 = Model(inputs, outputs)

model_1.compile(loss = 'mean_squared_error', 
              optimizer = SGD(lr=0.01), 
              metrics = ['accuracy'])
model_1.fit(x_train, y_train, 
          batch_size = 30, 
          epochs = 100, 
          verbose = 1);
predict = model_1.predict(np.array([4]))
print('f(26) = ', predict)

model_2.compile(loss='mean_squared_error', 
              optimizer='rmsprop', 
              metrics=['accuracy'])
model_2.fit(x_train, y_train, 
          batch_size = 10, 
          epochs = 100, 
          verbose = 1);
predict = model_2.predict(np.array([26]))
print('f(26) = ', predict)
```
